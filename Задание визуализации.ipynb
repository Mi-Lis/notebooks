```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100)

fig2, axs = plt.subplots(nrows=1, ncols = 1)

axs.plot(t, np.sin(t), linewidth=2, label = 'sin')
axs.legend(loc = 'best')
axs.tick_params(labelsize=16)
axs.set_xlabel('$t$', size = 16)
axs.xaxis.set_label_coords(1.05, -0.025)
axs.set_ylabel('$f(t)$',rotation=0, size = 16)
axs.yaxis.set_label_coords(-.1, .95)
axs.grid()

axs.plot(t, np.cos(t), '+r', linewidth=2, label = 'cos')
axs.legend(loc = 'best')
axs.tick_params(labelsize=16)
axs.set_xlabel('$t$', size = 16)
axs.xaxis.set_label_coords(1.05, -0.025)
axs.yaxis.set_label_coords(-.1, .95)
axs.grid()
fig2.savefig('f.png')
```
